# Ans.6 different pattern using nested while loop 

```python
# Right angeled Trianlge

n = int(input("Enter number of rows: "))
i = 1
while i<=n:
    j=1
    while j<=i:
        print("$",end="")
        j+=1
    print()
    i+=1
```

```python
# Inverted Right triangle
n = int(input('Enter number of rows : '))
 
i = 1
while i <= n :
    j = n
    while j >= i:
        print("*", end = " ")
        j -= 1
    print()
    i += 1
```

```python
# Number Pattern
n = int(input('Enter number of rows : '))
 
k = 1
i = 1
while i <= n :
    j = 1
    while j <= i:
        print("{:3d}".format(k), end = " ")
        j += 1
        k += 1
    print()
    i += 1
```
