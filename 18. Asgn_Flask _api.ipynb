# Q1. What is Flask Framework? What are the advantages of Flask Framework?
Flask Framework is a lightweight web application framework written in Python. It is designed to be simple, flexible, and easy to use, and provides a variety of features for building web applications, including routing, templating, and session management. Some advantages of Flask Framework are:

- Lightweight and easy to learn
- Flexible and customizable
- Built-in development server and debugger
- Large and active community
- Support for extensions and plugins


# Q3. What is App routing in Flask? Why do we use app routes?
App routing in Flask refers to the process of mapping URLs to view functions. We use app routes in Flask to define the URLs that our application should respond to and the corresponding functions that should be called when those URLs are requested. App routing helps in organizing the code and making it more modular.



# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
The url_for() function is used in Flask for URL building. It takes the name of a view function as its first argument and generates a URL based on the rules defined in the app's routing table. 

If we don't use url_for() function, we would have to manually construct the URLs in our templates, which can be tedious and error-prone. Additionally, if we ever change the URL structure of our application, we would have to go through all of our templates and update the URLs manually. By using the url_for() function, we can ensure that our URLs are always up-to-date and consistent throughout our application.
